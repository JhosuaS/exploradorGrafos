## 1. Introducción
En el área de las ciencias de la computación, las estructuras de datos avanzadas son muy importantes porque permiten representar y resolver problemas reales de manera eficiente [2]. En este proyecto se desarrolló un “Explorador Inteligente de Grafos”, cuyo objetivo es simular y optimizar rutas de evacuación dentro del campus de la Escuela Politécnica Nacional (EPN) [2].

Para representar el campus se utilizó un grafo no ponderado, donde cada nodo representa una facultad o punto seguro. El proyecto fue desarrollado aplicando Programación Orientada a Objetos (POO), implementando desde cero los algoritmos de Búsqueda en Anchura (BFS) y Búsqueda en Profundidad (DFS) [4].


## 2. Objetivos
Desarrollar una aplicación en Python utilizando Programación Orientada a Objetos (POO) que permita representar la infraestructura de la Escuela Politécnica Nacional (EPN) mediante un grafo, con el propósito de simular y analizar rutas de evacuación [5].

**Objetivos Específicos:**
* Representar el campus universitario mediante un grafo no ponderado, obteniendo la información desde un archivo JSON [6].
* Implementar desde cero los algoritmos BFS y DFS sin usar librerías externas [6].
* Comparar el funcionamiento de ambos algoritmos para determinar cuál ofrece la ruta más eficiente en una emergencia [6].


## 3. Definición del Problema
La EPN cuenta con un campus grande conformado por varias facultades. Debido al riesgo de sismos o incendios, el problema principal consiste en encontrar la manera más rápida y segura de evacuar a las personas hacia los puntos de encuentro [7, 8]. Se propone un sistema basado en grafos para calcular automáticamente rutas de evacuación y evitar congestiones [8].


## 4. Modelado del problema como grafo
Se utilizó un grafo no dirigido y no ponderado. Los nodos representan facultades y puntos de encuentro, y las aristas los caminos transitables. La información se almacena en `campus_map.json` como una lista de adyacencia [9].


## 5. Diseño de Clases
Las clases principales (`models.py`) son:
* **Clase Nodo:** Representa una ubicación. Almacena nombre, tipo, zona y un indicador `es_nodo_seguro` [10].
* **Clase Arista:** Representa la conexión entre un nodo origen y un destino [11].
* **Clase Grafo:** Almacena todos los nodos y conexiones, incluyendo el método `cargar_desde_diccionario` para leer el JSON [11].


## 6. Implementación de BFS y DFS
**6.1. Búsqueda en Anchura (BFS)**
Implementado con una lógica FIFO (Cola), recorre el mapa por niveles visitando primero los nodos más cercanos. Registra nodos para evitar ciclos y garantizar la ruta más corta [12, 13].

**6.2. Búsqueda en Profundidad (DFS)**
Implementado mediante recursividad (Pila LIFO). Explora un camino hasta el fondo y realiza backtracking si encuentra un camino sin salida. No garantiza la ruta más corta [13, 14].

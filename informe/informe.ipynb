## 1. Introducción
En el área de las ciencias de la computación, las estructuras de datos avanzadas son muy importantes porque permiten representar y resolver problemas reales de manera eficiente. Una de las estructuras más utilizadas son los grafos, ya que ayudan a representar conexiones entre distintos puntos, como redes de transporte, conexiones de internet o rutas de evacuación.
En este proyecto se desarrolló un “Explorador Inteligente de Grafos”, cuyo objetivo es simular y optimizar rutas de evacuación dentro del campus de la Escuela Politécnica Nacional (EPN). Esto es importante porque, en situaciones de emergencia como incendios o sismos, es necesario encontrar rutas rápidas y seguras para evacuar a estudiantes y docentes hacia zonas seguras.

Con estos algoritmos, el sistema puede encontrar rutas de evacuación y comparar cuál estrategia resulta más eficiente en una situación de emergencia. ## 3. Definición del Problema
La Escuela Politécnica Nacional (EPN), ubicada en la ciudad de Quito, cuenta con un campus grande conformado por varias facultades, laboratorios, edificios administrativos y zonas recreativas [@epn]. Todos los días, miles de estudiantes, docentes y personal administrativo circulan por estas instalaciones. Debido a la ubicación geográfica de Quito, existe un riesgo constante de eventos sísmicos. Además, en infraestructuras grandes también pueden ocurrir emergencias como incendios, fallas eléctricas o problemas estructurales. En este tipo de situaciones, muchas personas pueden desorientarse o entrar en pánico, provocando que se utilicen rutas inadecuadas y aumentando el tiempo de evacuación.
El problema principal consiste en encontrar la manera más rápida y segura de evacuar a las personas desde cualquier edificio del campus hacia los puntos de encuentro o zonas seguras establecidas por la universidad [@epn_mapa]. Si no existe una guía adecuada, pueden producirse congestiones en pasillos o caminos, dificultando aún más la salida de las personas.
Para solucionar este problema, se propone desarrollar un sistema inteligente basado en grafos que represente el mapa del campus universitario. De esta manera, el sistema podrá calcular automáticamente rutas de evacuación utilizando algoritmos de búsqueda, permitiendo encontrar el camino más corto y eficiente entre un punto de origen y una zona segura. Con este proyecto se busca reducir el tiempo de evacuación, mejorar la organización durante una emergencia y disminuir los riesgos para la comunidad universitaria.


## 4. Modelado del problema como grafo
Se utilizó un grafo no dirigido y no ponderado. Los nodos representan facultades y puntos de encuentro, y las aristas los caminos transitables. La información se almacena en `campus_map.json` como una lista de adyacencia. El grafo es no dirigido porque pueden existir rutas en ambos sentidos y además no ponderado por se representó a todas las aristas con el mismo valor o distancia.
En contraste, en otros escenarios sería necesario evaluar el peso asignado a las aristas para incluir la ruta más eficiente considerando la distancia o tiempo de recorrido desde el punto de peligro hasta el punto de evacuación.

## 6. Implementación de BFS y DFS
**6.1. Búsqueda en Anchura (BFS)**
Implementado con una lógica FIFO (*Cola*). El algoritmo recupera la información del nodo que se está visitando y sus vecinos, por lo que antes de avanzar averigua si alguno de los vecinos es en realidad un punto seguro [@stanford]. La implementación correspondiente se puede visualizar en `bfs.py`

**6.2. Búsqueda en Profundidad (DFS)**
Se implementó a través de recursividad mediante la pila de llamadas del sistema. El algoritmo recorre el grafo mientras pregunta si el primer vecino del nodo en revisión es un punto seguro, así hasta llegar al último vecino posible [@stanford]. Cada llamada regresa a la anterior mientras no se haya encontrado el punto seguro, así hasta haber preguntando por todos los puntos en el componente conectado.
Además, el algoritmo genera una nueva estructura denominada `arbol_dfs`, que es un diccionario en el que se identifican los padres de cada recorrido.
La implementación correspondiente se puede visualizar en `dfs.py`.
